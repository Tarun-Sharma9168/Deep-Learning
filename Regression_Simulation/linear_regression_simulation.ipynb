```python
#importing the python libraries
import pylab
import numpy
```

```python
#generating the numbers between -1 to +1
x=numpy.linspace(-1,1,100)
```

```python
x
```

```python
#generating signal function means the function what we want 
signal=2+x+2*x*x
#noise function is also there in which we are using the normal distribution of mean 0 and standard deviation 0.1
noise=numpy.random.normal(0,0.1,100)
```

```python
#observing the signal function
signal
```

```python
#observing the noise in the data
noise
```

```python
#Bringing Noise to the data and plot noise ,unnoised data and only noise which is there in 
y=signal+noise
pylab.plot(signal,'b')
pylab.plot(y,'g')
pylab.plot(noise,'r')
pylab.xlabel("x")
pylab.ylabel("y")
pylab.legend(["Without Noise","With Noise","Noise"],loc=2)
```

```python
#making training and testing data
x_train=x[0:80]
y_train=y[0:80]
```

```python
#Model with Degree 1
pylab.figure()
degree=2
X_train=numpy.column_stack([numpy.power(x_train,i) for i in range(0,degree)])
model=numpy.dot(numpy.dot(numpy.linalg.inv(numpy.dot(X_train.transpose(),X_train)),X_train.transpose()),y_train)
pylab.plot(x,y,'g')
pylab.xlabel("x")
pylab.ylabel("y")
predicted=numpy.dot(model,[numpy.power(x,i) for i in range(0,degree)])
pylab.plot(x,predicted,'r')
pylab.legend(["Actual","Predicted"],loc=2)
```

```python
#training RMSE (Root Mean Square Error)
train_rmse1=numpy.sqrt(numpy.sum(numpy.dot(y[0:80]-predicted[0:80],y_train-predicted[0:80])))
```

```python
#test RMSE (Root Mean Square Error)
test_rmse1=numpy.sqrt(numpy.sum(numpy.dot(y[80:]-predicted[80:],y[80:]-predicted[80:])))
```

```python
print("Train RMSE (Degree =1)",train_rmse1)
print("Test RMSE (Degree = 1)",test_rmse1)
```
